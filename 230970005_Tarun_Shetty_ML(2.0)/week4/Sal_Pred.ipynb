```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import re
from sklearn.preprocessing import LabelEncoder
from tabulate import tabulate
```

```python
df = pd.read_csv("../datasets/abalone.csv")
df[:5]
```

```python
le =LabelEncoder()
```

```python
df.Sex = le.fit_transform(df.Sex)
```

```python
df
```

```python
df[:5]
```

```python
print(df.corr())
```

```python
lr = LinearRegression()
```

```python
rings = df['Rings']
df = df.drop(columns=['Sex','Rings'])
x=df
```

```python
df.corr()
from sklearn.metrics import r2_score
```

```python
results = []
for split_ratio in [i/10 for i in range(1, 10)]:
    X_train, X_test, Y_train, Y_test =train_test_split(x,rings,test_size=split_ratio,random_state=42)
    lr.fit(X_train,Y_train)
    Y_predict=lr.predict(X_test)
    r2_s = r2_score(Y_test,Y_predict)
    results.append([split_ratio, r2_s])

print(tabulate(results, headers=["Train Ratio", "R2_Score"], tablefmt="pretty"))
```
