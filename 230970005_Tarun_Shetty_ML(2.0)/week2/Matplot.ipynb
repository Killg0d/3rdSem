```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,math.pi*2,0.5)
y=np.sin(x)
```

```python
plt.scatter(x,y)
plt.xlabel("angle")
plt.ylabel("sin of x")
plt.legend(['datapoints'])
for xy in zip(x,y):
    plt.annotate("(%.2f, %.2f)" % xy,xy=xy)
plt.savefig('sin.jpg')
```

```python
class_his={'CSE':33,'EC':44,'DSE':40}
courses=list(class_his.keys())
values=list(class_his.values())
print(courses)
```

```python
fig=plt.figure(figsize=(10,5))
plt.bar(courses,values,color="red",width=0.5)
plt.xlabel("Courses")
plt.ylabel("Students")
plt.title("Students in courses")
plt.legend("")
```
