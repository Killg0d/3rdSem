```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv("income.data.csv")
```

```python
data
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
plt.scatter(data['income'] ,data['happiness'])
```

```python
happiness_counts=data["happiness"].round().value_counts()

plt.figure(figsize=(8, 6))
plt.pie(happiness_counts.sort_index(),labels=happiness_counts.index,autopct='%1.1f%%')
plt.title('Distribution of Rounded Happiness Values')
plt.show()
```

```python
plt.hist(data["happiness"].round(), bins=range(int(data["happiness"].round().min()), int(data["happiness"].round().max()) + 2), edgecolor='black')
```

```python


plt.figure(figsize=(10, 5))

# First subplot for x
plt.subplot(1, 2, 1)
plt.boxplot(data["income"])
plt.title('Boxplot of income')

# Second subplot for y
plt.subplot(1, 2, 2)
plt.boxplot(data["happiness"])
plt.title('Boxplot of happiness')

plt.show()
```
