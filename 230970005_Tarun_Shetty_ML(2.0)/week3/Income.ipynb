```python
m=0
c=0
L=0.001 #learning rate
epochs = 1000
n=float(len(x))
print(len(x))


for i in range(epochs):
    y_pred = m*x +c #curr prediction
    D_m = (-2/n) * sum(x * (y - y_pred)) #derivatives wrt m
    D_c = (-2/n) * sum(y-y_pred) #derivatives wrt c
    m = m - L * D_m
    c = c - L * D_c
print(m,c)
```

```python
data.plot(kind='scatter',x='income',y='happiness')
Y_pred = m * x + c
print(Y_pred.shape)
plt.plot(x,Y_pred,color='red')
```
